```python
def read_input(fname):
    data = []
    data.append([])
    with open(fname, 'r') as inf:
        for line in inf.readlines():
            if line.strip() == '':
                continue
            data.append(['.'] + [c for c in line.strip()] + ['.'])
            
    data[0] = ['.' for _ in range(0, len(data[1]))]
    data.append(['.' for _ in range(0, len(data[1]))])
    
    return data
```

```python
def find_part_numbers(data):
    partnums = []
    for i_l, line in enumerate(data):
        i = 1
        while i < len(line):
            c = line[i]
            
            # Find a number
            if c.isdigit():
                success = False
                idx_start = i-1
                val = c
                while line[i+1].isdigit():
                    val += line[i+1]
                    i += 1
                idx_end = i+1
                # Now check if it is valid
                
                for row in range(i_l-1, i_l+2):
                    for col in range(idx_start, idx_end+1):
                        if data[row][col] != '.' and not data[row][col].isdigit():
                            success = True
                            partnums.append(int(val))
                            break
                    if success:
                        break
            i += 1
            
    return partnums
```

```python
def find_num(line, idx, direct):
    val = ''
    while line[idx+direct].isdigit():
        val += line[idx+direct]
        idx += direct
    if val != '':
        if direct < 0:
            return int(val[::-1])
        return int(val)
        
    return None
                
def find_gear_ratios(data):
    ratios = []
    for i_l, line in enumerate(data):
        for i, c in enumerate(line):
            # Find a *
            if c == '*':
                numbers = []
                # Found a gear
                
                for d in (-1, 1):
                    num = find_num(line, i, d)
                    if num is not None:
                        numbers.append(num)
                            
                for l in (-1, 1):
                    # First check that directly above below is empty
                    if data[i_l+l][i].isdigit():
                        # We have a number directly above, find the start
                        idx = i-1
                        while data[i_l+l][idx].isdigit():
                            idx -= 1
                        num = find_num(data[i_l+l], idx, 1)
                        numbers.append(num)
                    else:
                        # Diagonal cases
                        for idx in (-1, 1):
                            num = find_num(data[i_l+l], i, idx)
                            if num is None:
                                continue
                            numbers.append(num)

                if len(numbers) == 2:
                    ratios.append(numbers[0] * numbers[1])
            
    return ratios
```

```python
print('Test case\n')

data = read_input('input3a.txt')

vals = find_part_numbers(data)
res = sum(vals)
print(f'Sum of part numbers {res}')

assert res == 4361

print('Puzzle case\n')

data = read_input('input3.txt')

vals = find_part_numbers(data)
res = sum(vals)
print(f'Sum of part numbers {res}')

assert res == 546563

print('\n****\nSecond puzzle\n****\n')

data = read_input('input3a.txt')

vals = find_gear_ratios(data)
res = sum(vals)
print(f'Sum of gear ratios {res}')

assert res == 467835

print('Puzzle case\n')

data = read_input('input3.txt')

vals = find_gear_ratios(data)
res = sum(vals)
print(f'Sum of gear ratios {res}')

assert res == 91031374
```
