```python
import numpy as np
```

```python
def read_input(fname):
    patterns = []
    with open(fname, 'r') as inf:
        data = []
        for line in inf.readlines():
            if line.strip() == '':
                patterns.append(np.array(data))
                data = []
                continue
            data.append([c for c in line.strip()])
    return patterns
```

```python
def find_mirrors(patterns):
    val = 0
    for data in patterns:
        len_y, len_x = data.shape

        # Columns
        for x in range(0, len_x-1):
            min_x = max(0, 2*x-len_x+2)
            max_x = min(len_x, x*2+2)
        
            if np.all(data[:,min_x:x+1] == data[:,x+1:max_x][:,::-1]):
                val += x+1

        # Rows
        for y in range(0, len_y-1):
            min_y = max(0, 2*y-len_y+2)
            max_y = min(len_y, y*2+2)
        
            if np.all(data[min_y:y+1,:] == data[y+1:max_y,:][::-1,:]):
                val += 100*(y+1)

    return val
            
```

```python
def find_smudged_mirrors(patterns):
    val = 0
    for data in patterns:
        len_y, len_x = data.shape
        n_elem = len_x * len_y

        # Columns
        for x in range(0, len_x-1):
            min_x = max(0, 2*x-len_x+2)
            max_x = min(len_x, x*2+2)
        
            if np.sum(data[:,min_x:x+1] != data[:,x+1:max_x][:,::-1]) == 1:
                val += x+1

        # Rows
        for y in range(0, len_y-1):
            min_y = max(0, 2*y-len_y+2)
            max_y = min(len_y, y*2+2)
        
            if np.sum(data[min_y:y+1,:] != data[y+1:max_y,:][::-1,:]) == 1:
                val += 100*(y+1)

    return val
            
```

```python
print('*****\nPuzzle1\n*****\n')

print('Test case\n')

data = read_input('input13a.txt')

sumval = find_mirrors(data)
    
print(f'Sum is {sumval}')

assert sumval == 405

print('\nPuzzle case\n')

data = read_input('input13.txt')

sumval = find_mirrors(data)
    
print(f'Sum is {sumval}')

assert sumval == 31956

print('\n*****\nPuzzle2\n*****\n')

print('Test case\n')

data = read_input('input13a.txt')

sumval = find_smudged_mirrors(data)
    
print(f'Sum is {sumval}')

assert sumval == 400

print('\nPuzzle case\n')

data = read_input('input13.txt')

sumval = find_smudged_mirrors(data)
    
print(f'Sum is {sumval}')

assert sumval == 37617

```
