```python
import copy
```

```python
def map_values(invals, inf):
    outvals = [i for i in invals]
    line = inf.readline().strip()
    while line != '':
        outstart, instart, length = (int(i) for i in line.split())
        for i, v in enumerate(invals):
            if v >= instart and v < instart+length:
                outvals[i] = v + (outstart-instart)
        line = inf.readline().strip()
    return outvals
```

```python
def map_values_range(inranges, inf):
    outranges = []
    next_ranges = copy.deepcopy(inranges)
    line = inf.readline().strip()
    while line != '':
        out_start, in_start, length = (int(i) for i in line.split())
        in_end = in_start + length - 1
        offset = out_start - in_start
                
        check_ranges = copy.deepcopy(next_ranges)
        next_ranges = []
        for i, v in enumerate(check_ranges):
            start, end = v
            
            if start < in_start:
                if end < in_start:
                    next_ranges.append((start, end))
                if end >= in_start:
                    next_ranges.append((start, in_start-1))
                    outranges.append((out_start, min(end, in_end)+offset))
                if end > in_end:
                    next_ranges.append((in_end+1, end))
            elif start >= in_start and start <= in_end:
                outranges.append((start+offset, min(end, in_end)+offset))
                if end > in_end:
                    next_ranges.append((in_end+1, end))
            else:
                next_ranges.append((start, end))
        line = inf.readline().strip()
        
    return outranges + next_ranges
```

```python
def read_and_map(fname):
    with open(fname, 'r') as inf:
        line = inf.readline()
        seeds = [int(i) for i in line.strip().split()[1:]]
        # Eat empty line

        inf.readline()
        
        assert inf.readline().strip() == 'seed-to-soil map:'
        
        soils = map_values(seeds, inf)
        
        assert inf.readline().strip() == 'soil-to-fertilizer map:'
        
        fertilizer = map_values(soils, inf)
        
        assert inf.readline().strip() == 'fertilizer-to-water map:'
        
        water = map_values(fertilizer, inf)
        
        assert inf.readline().strip() == 'water-to-light map:'
        
        light = map_values(water, inf)
        
        assert inf.readline().strip() == 'light-to-temperature map:'
        
        temp = map_values(light, inf)
        
        assert inf.readline().strip() == 'temperature-to-humidity map:'
        
        humid = map_values(temp, inf)
        
        assert inf.readline().strip() == 'humidity-to-location map:'
        
        location = map_values(humid, inf)
        
    return set(location)
```

```python
def read_and_map_ranges(fname):
    with open(fname, 'r') as inf:
        line = inf.readline()
        vals = [int(i) for i in line.strip().split()[1:]]
        seeds = [(vals[i], vals[i]+vals[i+1]-1) for i in range(0, len(vals), 2)]
        # Eat empty line

        inf.readline()
        
        assert inf.readline().strip() == 'seed-to-soil map:'
        
        soils = map_values_range(seeds, inf)
        
        assert inf.readline().strip() == 'soil-to-fertilizer map:'
        
        fertilizer = map_values_range(soils, inf)
        
        assert inf.readline().strip() == 'fertilizer-to-water map:'
        
        water = map_values_range(fertilizer, inf)
        
        assert inf.readline().strip() == 'water-to-light map:'
        
        light = map_values_range(water, inf)
        
        assert inf.readline().strip() == 'light-to-temperature map:'
        
        temp = map_values_range(light, inf)
        
        assert inf.readline().strip() == 'temperature-to-humidity map:'
        
        humid = map_values_range(temp, inf)
        
        assert inf.readline().strip() == 'humidity-to-location map:'
        
        location = map_values_range(humid, inf)
        
    return location
```

```python
print('*****\nPuzzle1\n*****\n')

print('Test case\n')

res = min(read_and_map('input5a.txt'))

print(f'Location is {res}')

assert res == 35

print('\nPuzzle case\n')

res = min(read_and_map('input5.txt'))

print(f'Location is {res}')

assert res == 535088217

print('\n*****\nPuzzle2\n*****\n')

print('Test case\n')

res = min(read_and_map_ranges('input5a.txt'))[0]

print(f'Location is {res}')

assert res == 46

print('\nPuzzle case\n')

res = min(read_and_map_ranges('input5.txt'))[0]

print(f'Location is {res}')

assert res == 51399228
```
